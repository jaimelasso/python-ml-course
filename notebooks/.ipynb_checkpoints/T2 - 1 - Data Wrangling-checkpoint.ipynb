### Generación de números aleatorios 

```python
import numpy as np
```

```python
# Generar un número aleatorio entre 1 y 100
np.random.randint(1,100)
```

```python
# Forma clásica de generar un número aleatorio entre 0 y 1 (con decimales)
np.random.random()
```

```python
# Método que genera una lista de n números aleatorios dentro del intervalo [a,b]
def randint_list(n, a, b):
    x = []
    for i in range(n):
        x.append(np.random.randint(a,b))
    return x
```

```python
randint_list(11, 1, 100)
```

```python
import random #Haciendo lo mismo, pero con esta librería
```

```python
for i in range(10):
    print(random.randrange(0,100,7)) #Imprimiendo los 10 primeros números aleatorios múltiplos de 7
```

## Seeds: Nos permite elegir el número de partida para la generación de # aleatorios

Esto se hace con el propósito de que al momento de modelar podamos hacer comparables los resultados entre equipos de trabajo diferente. Importante para la reproductibilidad del experimento a posteriori.

```python
np.random.seed(2018)
for i in range(10):
    print(np.random.random())
```
